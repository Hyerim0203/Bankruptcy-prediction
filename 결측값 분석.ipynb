## 회사 설립년도 구하기

```python
zero= np.load('C:\\Users\\이혜림\\Desktop\\신용도분석\\Zeromatrix_save\\Zeromatrix_save.npy') # 데이터 불러오기
zero # 0 이 처음으로 나오는 곳이 회사 설립날짜(회사가 존재하는 년도)
```

```python
company_start_year = [list(range(0,2191)),[]]

for idx, company in enumerate(zero):
    start_year = 2000 + company.argmin() # 0이 처음으로 나오는 index + 2000 으로 회사가 처음으로 존재한 년도 구함
    company_start_year[1].append(start_year)
    
company_start_year = pd.DataFrame(company_start_year).T
company_start_year.columns = ["company","start_year"]
```

```python
company_start_year["start_year"].value_counts()
```
