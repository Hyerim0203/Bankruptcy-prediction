## 전처리

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv("./all.csv")
data=data.iloc[:,1:]
```

```python
# Stock, COE, CSE칼럼들 삭제
for i in data.columns:
    if "Stock" in i:
        data=data.drop([i], axis=1)
    if "COE" in i:
        data=data.drop([i], axis=1)
    if "CSE" in i:
        data=data.drop([i], axis=1)
        
```

```python
raw_data = data.copy()
```

# 데이터 년도별로

```python
# 년도별 딕셔너리
data_year = {idx:pd.DataFrame() for idx in range(2000,2018)}    

for year in range(2000,2018):
    for i in data.columns:
        if str(year) in i:
            data_year[year][i]=data[i]
```

# 데이터 변수별로

```python
data_col = {i[4:]:pd.DataFrame() for i in data_year[2000]}

for i in data.columns:
    for j in data_col:
        if j in i:
            data_col[j][i] = data[i]
```

# 칼럼 형식 변환

## ao

```python
# 총 ao 라벨 값
ao_value = []
for i in range(2000,2018):
    uni = data_year[i]["{}AO".format(i)].unique()
    for u in uni:
        if u not in ao_value:
            ao_value.append(u)
            
print(ao_value)
```

```python
ao_label = {"적정":0,"한정":1,"부적정":2,"의견거절":3,"감사미필":4,"0":4}
for col in data.columns:
    if "AO" in col:
        for idx, i in enumerate(data[col]):
            for col_idx, j in enumerate(ao_label):
                if j in str(i):
                    data[col][idx]=ao_label[j]
```

## ms

```python
for i in range(len(data['MS'])):
    if data['MS'][i]=='코스닥상장':
        data['MS'][i]=0
    elif data['MS'][i]=='상장':
        data['MS'][i]=1
    elif data['MS'][i]=='코넥스':
        data['MS'][i]=2
    elif data['MS'][i]=='상장관리':
        data['MS'][i]=3
    elif data['MS'][i]=='코스닥관리':
        data['MS'][i]=4    
```

## cs

```python
for i in range(len(data['CS'])):
    if data['CS'][i]=='대기업':
        data['CS'][i]=0
    elif data['CS'][i]=='중소기업':
        data['CS'][i]=1
    elif data['CS'][i]=='기타':
        data['CS'][i]=2
```

```python
data.to_csv("./all_preprocessing.csv", index=False)
```

### 년도 및 gdp 칼럼 추가

```python
# gdp 데이터 불러오기
gdp = pd.read_csv("./data/gdp_csv.csv",sep="\t")
gdp.columns = ["YEAR","GDP"]
gdp["YEAR"][1]=2001

# gdp 값 float형 변환
gdp["GDP"]=gdp["GDP"].str.replace(",","")
gdp["GDP"]=gdp["GDP"].apply(lambda x: float(x))
```

```python
sero_data = pd.read_csv("./data/sero.csv")
sero_data["gdp"]=0 # gdp 칼럼 추가

# sero data에 year one-hot 칼럼추가 및 gdp 칼럼 값 업데이트
for i in range(0,19):
    # 각 년도 칼럼 추가
    year = 2000+i
    sero_data[str(year)]=0
    
    # 해당 년도, gdp 값 업데이트
    sero_data[str(year)][2191*i:2191*(i+1)]=1
    value = gdp.loc[gdp["YEAR"]== year,"GDP"][i] # 해당 년도 gdp 값
    sero_data["gdp"][2191*i:2191*(i+1)] = value
```

```python
sero_data.to_csv("./data/sero_data.csv", index=False)
```

### 텍스트 데이터 - onehot encoding
- label encoding 된 것들 : AO, CS, MS

```python
# 텍스트 데이터 one-hot
sero_data = pd.read_csv("./data/sero_data.csv")
```

```python
for col in ["AO","CS","MS"]:
    a = pd.get_dummies(sero_data[col]) # one hot encoding
    a.columns = [f"{col}{i}" for i in range(a.shape[1])] # 칼럼명 생성
    sero_data = sero_data.drop([col],axis=1) # 기존 칼럼 삭제
    sero_data = pd.concat([sero_data,a],axis=1) # 새로운 칼럼 추가
```

```python
sero_data.to_csv("./data/sero_data.csv", index=False)
```
